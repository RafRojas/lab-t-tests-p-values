## Lab | Inferential statistics - T-test & P-value:

### Instructions:

1) One tailed t-test - In a packing plant, a machine packs cartons with jars. It is supposed that a new machine will pack faster on the average than the machine currently used. To test that hypothesis, the times it takes each machine to pack ten cartons are recorded. The results, in seconds, are shown in the tables in the file files_for_lab/machine.txt. Assume that there is sufficient evidence to conduct the t test, does the data provide sufficient evidence to show if one machine is better than the other?


```python
import pandas as pd
import numpy as np
import scipy.stats as st
```

```python
df = pd.read_csv('files_for_lab/machine.txt', encoding='UTF-16', sep = '\t')
```

```python
df.columns = ['New_machine', 'Old_machine']

df
```

μ: Number of seconds it takes a machine to pack 10 cartons.

H0: New machine's μ =< Old machine's μ    ---    New machine packs cartons as fast or faster than the old one.

H1: New machine's μ > Old machine's μ     ---    New machine packs cartons slower than the old one.

α: 0.05

```python
p_value = st.ttest_ind(df['New_machine'], df['Old_machine'], equal_var=False)
print(p_value)
```

The p value is lower than 0.05, we do not reject H0. The new machine's average time packing is faster than the old machine's.

2) Matched Pairs Test - In this challenge we will compare dependent samples of data describing our Pokemon (file files_for_lab/pokemon.csv). Our goal is to see whether there is a significant difference between each Pokemon's defense and attack scores. Our hypothesis is that the defense and attack scores are equal. Compare the two columns to see if there is a statistically significant difference between them and comment your result.

H0: Attack = Defense    ---    Attack is equal to defense.

H1: Attack != Defense   ---    Attack and defense are different.

α: 0.05

```python
pokemon = pd.read_csv('files_for_lab/pokemon.csv')

pokemon.head()
```

```python
p_value = st.ttest_ind(pokemon['Attack'], pokemon['Defense'], equal_var=False, alternative='two-sided')
print(p_value)
```

The p value is lower than our 0.05 alpha, meaning it is in the rejection zone of the left tail of the distribution. We reject H0, there's a statistically significative difference between attack and defense.
